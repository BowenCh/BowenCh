```python
def delet_nest(y:dict):
    yy = {}
    x = y.copy()
    if x['linkType'] == 'UAT':
        x['alt']['geometric'] = ''
        x['acID'] = ''
        
    elif 'acID' not in list(x.keys()):
        x['acID'] = ''
        
        
    for i in y.keys(): #按keys 遍历
        if type(y[i]) == dict: #如果某一key 对应value是dict
            yy[i] = y[i]          
    for j in yy.keys():
        del x[j]  
    for m in yy.values():
        for n in m.keys():
            x[n] = m[n]
            
    #headers = ['linkType', 'TimeSinceMidnight', 'acID', 'address', 'type', 'lat', 
    #          'lon', 'pressure', 'geometric', 'speed', 'track', 'state']
    
    #ss = (set(headers).difference(set(list(x))))
    
    #if len(ss):
    #   for b in ss:
    #       x[b] =''
        
    return x
```
